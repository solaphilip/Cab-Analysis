```python
df['Date of Travel'] = pd.to_datetime(df['Date of Travel'])

# Find the date with the highest number of travels
highest_travel_date = df['Date of Travel'].value_counts().idxmax()
print(f"Highest travel date: {highest_travel_date}")

# Plot the count of travels by date and company
sns.countplot(x='Date of Travel', hue='Company', data=df)
plt.legend(handletextpad=1.0, borderaxespad=1.0)
plt.title('Customer by Company and Travel Date')
plt.show()
```
